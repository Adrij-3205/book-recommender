```python
books=pd.read_csv("Books.csv")
```

```python
books.columns
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
books.head()
```

```python
users=pd.read_csv('Users.csv')
users.head()
```

```python
ratings=pd.read_csv('Ratings.csv')
ratings.head()
```

```python
books.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
x=ratings['User-ID'].value_counts()>150
x[x].shape
```

```python
y=x[x].index
y
```

```python
ratings=ratings[ratings['User-ID'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('Book-Title')['Book-Rating'].count().reset_index()
number_rating.rename(columns={'Book-Rating':'Number of Ratings'},inplace=True)
number_rating
```

```python
final_rating=ratings_with_books.merge(number_rating,on='Book-Title')
final_rating
```

```python
final_rating=final_rating[final_rating['Number of Ratings']>=40]
final_rating.shape
```

```python
final_rating.drop_duplicates(['User-ID','Book-Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
np.where(book_pivot.index=='Harry Potter and the Prisoner of Azkaban (Book 3)')[0][0]
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[412,:].values.reshape(1,-1),n_neighbors=6)
```
